```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Downloads/companies.csv')
df.head()
```

```python
# Question 1 :  count of how many of the companies in the data set are from each state. 

sns.countplot(x='state', data=df)
plt.xticks(rotation = 90)
plt.title('Count of Companies by State')
plt.show()
```

```python
# Question 3: create a bar chart that shows a sum of how many total job openings there are in each industry.

sns.barplot(y='Industry', x='job_openings', data=df, errorbar=None, estimator = sum)
plt.title('Total Job Openings per Industry')
plt.show()
```

```python
# Question 5: create a line chart that shows the average growth rate for each founded year.

sns.lineplot(x='founded', y='growth_percentage', data=df)
plt.xlabel('Year Founded')
plt.ylabel('Growth Rate')
plt.title('Average Growth Rate per Year Founded')
plt.show()
```

```python
# Question 7: create a scatter plot that compares the number of employees to estimated revenues for each company.

sns.scatterplot(y='estimated_revenues', x='employees', data=df, color = 'green')
plt.title('Comparison of Estimated Revenues to Employees')
plt.show()
```

```python
# Question 9: create a histogram that measures the number of companies founded in each year.

sns.histplot(x='founded', data=df, bins=30)
plt.title('Count of Companies Founded')
plt.xlabel('Years Founded')
plt.show()
```
