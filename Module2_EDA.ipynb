```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\ryane\Downloads\housing.csv")
df.head()
```

```python
# Question 1: Create a scatter plot that shows land size on the x-axis and price on the y-axis. 
# Change the color of the points to reflect the type of the property.
sns.scatterplot(x='Landsize', y='Price', data=df, hue='Type')
plt.ylabel('Millions')
plt.title('Price vs Land Size')
plt.show()
```

```python
# Question 3: Create a line chart that shows the year built along the x-axis and the average price per year on the y-axis. 
# You will need to use aggregation and group by to accomplish this.

df_yearbuilt = df.groupby('YearBuilt')['Price'].mean().reset_index()

df_yearbuilt.head()
```

```python
sns.lineplot(x='YearBuilt', y='Price', data=df_yearbuilt)
plt.xlabel('Year Built')
plt.ylabel('Price (In Millions)')
plt.title('Average Price per Year Built')
plt.show()
```

```python
# Question 5: Create a bar chart that shows the region on the x-axis and the average price of the property on the y-axis. 
# You will need to use aggregation and group by to accomplish this.

df_region = df.groupby('Regionname')['Price'].mean().reset_index()
df_region.head()
```

```python
sns.barplot(x='Regionname', y='Price', data=df_region)
plt.xticks(rotation = 45)
plt.xlabel('Region')
plt.title('Average Price per Region')
plt.show()
```

```python
# Question 7: Create a histogram of property prices.

sns.histplot(x='Price', data=df, bins=30)
plt.title('Price Distribution')
plt.xlabel('Price (In Millions)')
plt.show()
```

```python
# Question 9: Create a box plot that shows the years that each property was built on the x-axis and the region on the y-axis. 
# These box plots should be oriented horizontally.

sns.boxplot(x='YearBuilt', y='Regionname', data=df, orient='h', hue = 'Regionname')
plt.title('Years Properties were Built')
plt.show()
```
